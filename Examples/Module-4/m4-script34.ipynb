```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Basic plots in Matplotlib

# x-axis values 
x = [5, 3, 8, 9, 1] 

# Y-axis values 
y = [6, 4, 9, 4, 2] 

# Function to plot 
plt.plot(x,y) 

# Function to show the plot 
plt.show()
```

```python
# Plotting Multiple Sets of Data
x = np.arange(1,5)
y = x**5
plt.plot([10,20,30,40],[100,200,300,400],'go', x,y,'r^')
plt.title("First Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# Bar plot

# x-axis values 
x = [5, 2, 9, 4, 7] 

# Y-axis values 
y = [10, 5, 8, 4, 2] 

# Function to plot the bar 
plt.bar(x,y) 

# function to show the plot 
plt.show() 
```

```python
# Histogram

# Y-axis values 
y = [10, 5, 8, 4, 2] 

# Function to plot histogram 
plt.hist(y) 

# Function to show the plot 
plt.show() 
```

```python
# Scatter Plot

# x-axis values 
x = [5, 2, 9, 4, 7] 

# Y-axis values 
y = [10, 5, 8, 4, 2] 

# Function to plot scatter 
plt.scatter(x, y) 

# function to show the plot 
plt.show() 
```

```python
# Pie Charts

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
pro_langs = ['c','c++','java', 'python','php']
stud = [23,14,32,50,41]
ax.pie(stud,labels=pro_langs,autopct="%1.2f%%")
plt.show()
```
